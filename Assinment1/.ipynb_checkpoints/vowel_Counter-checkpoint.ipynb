```python
#wap to ask user to enter the stritn and the vowel in it 
print("-------VOWEL COUNTER-------")
dummyString = input("Enter the string to count vowels in it: ")
count =0
#while('a','e' , 'i', 'o' ,'u' in dummyString.lower()):
   

for x in dummyString.lower():
    if x == 'a' or x =='e' or x == 'i'or x == 'o' or x =='u': #COMPARE VALUE OF X WITH EACH OF THE VOWELS CHARACTER
         count += 1 #COUNTS THE NUMBER OF VOWELS
print(f"{dummyString} contains {count} vowels")

```

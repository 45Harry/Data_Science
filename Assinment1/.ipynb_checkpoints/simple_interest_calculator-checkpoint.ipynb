```python
#WAP TO CALCULTE THE SIMPLE INTEREST TAKING INPUT FROM USER
print("WEL-COME TO SIMPLE INTEREST CALCULATER")
principle = int(input("Enter the Principle: "))
rate = int(input("Enter the Rate: "))
time = int(input("Enter the Time: "))
simpleInterest = (principle*time*rate)/100
print(f"Congrulation!!! Your Simple Interest is {simpleInterest}")
```

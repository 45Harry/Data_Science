```python
#wap to check for the leap year
print("******LEAP YEAR CHECKER*****")
yearInDays = int(input("Please enter the years: "))
if (yearInDays % 4 ==0 and not yearInDays /100) or yearInDays % 400==0 :
    print(f"{yearInDays} is leap year")
else: 
    print(f"{yearInDays} is not a leap year")
    
```

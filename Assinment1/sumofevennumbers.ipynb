```python
#wap to find the sum of even numbers
print("******Even Number Adder*****")
end = int(input("Enter the number upto where you want to add even number: "))
sumOfEven = 0
for x in range(1,end):
    if x % 2 == 0:
        sumOfEven += x
        print(sumOfEven)
print(f"The sum of even number upto {end} is ",sumOfEven)
    
        
```

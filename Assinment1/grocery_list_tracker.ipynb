```python
#wap to track the grocery list 
print("---------Grocery List Tracker----------")
nameOfItems = [] #empty list to store name of items
priceOfItems = [] #empty list to store price of items
sumAmount =0

for i in range(5):
    nameOfItems.append(input("Enter the name of a Items: ")) #add items to the list
    priceOfItems.append(int(input("Enter the price of a Items: "))) #add price to the pricelist

for x in priceOfItems:
    sumAmount = x + sumAmount #calculate the total amount 
    
print("Your Grocery are")
for y in nameOfItems:
    print(y)
print(f"The total amount is {sumAmount}") #print the total amount
    
```

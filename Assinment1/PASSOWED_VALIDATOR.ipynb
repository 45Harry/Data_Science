```python
#wap to check if the password entered by user is valid or not
#password must contain one number one letter and must be 8 char long
print("***********Password Checker***********")
passWord = input("Your password must contains one number and one uppercase letter and 8 charlong \n Enter your Password:")
upper = False
digit = False
if len(passWord) >=8:
    for x in passWord:
        if x.isupper():
            upper = True
            
        if x.isdigit():
            digit = True
    if upper == True and digit == True:
        print("Your Password is Strong :)")
    else:
        print("Please enter strong password")
            
        
     
else:
    print("Invalid Password!!! Your password must contains one number and one uppercase letter and 8 charlong")
```

```python
print("------------------ExTracker-----------------")
print("----Track Your Expense On The Go With US----")

myExpItems = [] # an empty list to add items or differet category
priceOfItems=[] # an empty list to add expenses for different category
more = True

totalBudget = int(input("Enter your total budget for a month: "))
print("Now Enter the Category and Expenses")

while more == True:
    myExpItems.append(input("Please enter the itmes name : "))
    priceOfItems.append(int(input("Please enter the price of an items: ")))

    choice = input("Want to add more items(Y/N) :) ")
    if choice.upper() == 'Y':
        more = True
    elif choice.upper() =='N':
        more = False
    else:
        print("Invalid Choice Please choose between Y/N")
        more = False
```

```python
itemsBook = dict(zip(myExpItems,priceOfItems)) # this will add first array as a key and second array as an items


for key,value in itemsBook.items():
    print(f"Your expenses for {key} is {value}")

totalExpenses = 0
for prce in priceOfItems:
    totalExpenses +=prce
print("Your total budget for a month is: ",totalBudget)
print("Your total expenses is : ",totalExpenses)
print("Your total saving in a month is: ", totalBudget-totalExpenses)
if totalBudget <= totalExpenses:
    print("You really need to cut some expenses and do some saving:( ")
    
    
```

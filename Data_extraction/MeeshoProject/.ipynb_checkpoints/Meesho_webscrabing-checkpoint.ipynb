```python
import pandas as pd
import requests as r
from bs4 import BeautifulSoup
from datetime import datetime
import os
```

```python
#working shell 
#url='''https://www.meesho.com/home-decor/pl/3tl'''
url= """https://www.meesho.com/belts-men/pl/3nn"""
url= """https://www.meesho.com/skirts-ladies/pl/3jf"""
url= """https://www.meesho.com/women-kurtis/pl/3j0"""

def soup_meesho(url):
    response = r.get(url)
    html_data = response.content
    soup = BeautifulSoup(html_data,'lxml')
    
    all_class = ['sc-dkrFOg ProductList__GridCol-sc-8lnc8o-0 cokuZA eCJiSA','sc-dkrFOg ProductList__GridCol-sc-8lnc8o-0 cokuZA jGAjKy',
                 'sc-dkrFOg ProductList__GridCol-sc-8lnc8o-0 cokuZA bGpzcx','sc-dkrFOg ProductList__GridCol-sc-8lnc8o-0 cokuZA ihwYlH']
    
    
    #print(file_name)
    
    
    Rating = []
    Products = []
    Price =[]
    Reviews =[]
    off_pri=[]
    dis =[]
    dilev_charg = []
    
    
    for cls in all_class:
        for item in soup.find_all('div',{'class':cls}):
          #  print(item.text.strip())
            scrap_data = item.text.strip().split('₹')
            
            if len(scrap_data)==2:
                prod, other_details = item.text.strip().split('₹')
               # print(other_details)
                rest_d,rat_rev=other_details.split(' Free Delivery')
                ##print(rat_rev)
                pric = rest_d
               # print(pric)
        
                Products.append(prod)
                Price.append(pric)
                off_pri.append(pric)
                dis.append(0)
                dilev_charg.append(0)
                if 'supplier' in rat_rev.lower():
                    Rating.append(rat_rev[:3])
                    Reviews.append(0)
                    
                elif 'reviews' in rat_rev.lower():
                    #print(rat_rev)
                    rat_re,_ = rat_rev.split(' ')
                   #print(rat_re)
                    Rating.append(rat_re[:3])
                    Reviews.append(rat_re[3:])
                else:
                    Rating.append(0)
                    Reviews.append(0)
                    
            elif len(scrap_data)==3:
                #print(scrap_data)
                prod,offered_price,rest_details = scrap_data
                Products.append(prod)
                off_pri.append(offered_price)
                pprice_dic,rate = rest_details.split(' offFree Delivery')
                Price.append(pprice_dic[:3])
                dis.append(pprice_dic[3:])
                dilev_charg.append(0)
                ret_rev = scrap_data[2].split(" offFree Delivery")[1]
                rating = ret_rev.split(" ")
                
                Reviews.append(rating[0][3:])
                #Reviews.append(scrap_data[2].split(" offFree Delivery")[1][3:])
                Rating.append(rating[0][:3])
                #print(rest_details)
                #print(offered_price)
            else:
                #print(scrap_data)
                Products.append(scrap_data[0])
                off_pri.append(0)
                Price.append(scrap_data[1].split(" ")[0])
                dilev_charg.append(scrap_data[2])
                dis.append(0)
                rat_revies = scrap_data[3].split(" ")
                Rating.append(rat_revies[0][2:5])
                Reviews.append(rat_revies[0][5:])
                #Reviews.append(scrap_data[4].split(" ")[3:6]))
                #print(prod)
                #print(offered_price)
                #print(Rating)
                #print(Reviews)
                    
                    
                
                      
    """print(len(Rating))
    print(len(Products))
    print(len(Price))
    print(len(Reviews)) 
    print(len(off_pri))
    print(len(dilev_charg))
    print(len(dis))"""
    
    all_data = {'Scrap_time':datetime.now(),'Product_Name':Products,'Offered_Price':off_pri,'Price':Price,'Discount':dis,'Rating':Rating,'Reviews':Reviews,'Delever_charge':dilev_charg}
    
    #defining file name from the url 
    file_name = url.split("https://www.meesho.com/")[1].split("/")[0]
    
    table_new = pd.DataFrame(all_data)
    
    #print(df)
    folder_finders = os.listdir('/home/harry/Documents/Code/MINDRISERS_DATA_SCIENCE/Data_Science/Data_extraction/MeeshoProject')
    
                                
    if f'{file_name}.csv' in folder_finders:
        table_old = pd.read_csv(f"/home/harry/Documents/Code/MINDRISERS_DATA_SCIENCE/Data_Science/Data_extraction/MeeshoProject/{file_name}.csv")
        final_table = pd.concat([table_old,table_new])
        final_table.to_csv(f'//home//harry//Documents//Code//MINDRISERS_DATA_SCIENCE//Data_Science//Data_extraction//MeeshoProject//{file_name}.csv',index = False)
        print("Data Appended to the existing Table")
    
    else:
        table_new.to_csv(f'//home//harry//Documents//Code//MINDRISERS_DATA_SCIENCE//Data_Science//Data_extraction//MeeshoProject//{file_name}.csv',index = False)
        print("Data Extracted sucessfully")
    #return df 
    return table_new


```

```python
url='''https://www.meesho.com/home-decor/pl/3tl'''
url="""https://www.meesho.com/lehengas/pl/3l6"""

soup_meesho(url)
```

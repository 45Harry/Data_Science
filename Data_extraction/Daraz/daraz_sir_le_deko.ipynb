```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Use the Service class for driver setup
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# Open the webpage
url = "https://www.daraz.com.np/catalog/?spm=a2a0e.pdp_revamp.search.3.5b9b4e336etQR4&q=iphone+15+pro+max&from_searchbox_hotwords=1&clickTrackInfo=matchType--99___wh_pid--___type--cli___scene--diwen___bucket--0___spm_id--hint.hp.sbox___result--iphone%2015%20pro%20max___textId--375550417040772256___score--0.748537___itemId--5___abtest--___pvid--66f7fb1c-6a7d-4493-bfcd-e105d3fcd0a3___scoreChain--0.000749,0.000000,0.000000___scm--1007.25945.379337.0___ifMannual--N"
driver.get(url)

# Allow time for the page to load
driver.implicitly_wait(10)

# Locate product elements using class name
product_elements = driver.find_elements(By.CLASS_NAME, "Bm3ON")

# Extract product names
brands = []
for element in product_elements:
    try:
        name_element = element.find_element(By.CLASS_NAME, "RfADt")  # Adjust if necessary
        brands.append(name_element.text.strip())  # Clean up the text
    except:
        brands.append(None)  # Handle missing names gracefully

# Locate the price elements by class name
price_elements = driver.find_elements(By.CLASS_NAME, "ooOxS")

# Extract the text of each price
prices = [element.text.strip() for element in price_elements]

# Quit the browser
driver.quit()

# Save data to a pandas DataFrame
data = {
    "Product Name": brands,
    "Price": prices
}
df = pd.DataFrame(data)

# Export to CSV
csv_filename = "smart_watch_data.csv"
df.to_csv(csv_filename, index=False)

print(f"Data saved to {csv_filename}")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
from selenium.webdriver.chrome.options import Options

options = Options()
options.add_argument("--headless")

# Setup WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# Base URL (replace `page=1` dynamically)
base_url = "https://www.daraz.com.np/catalog/?page={page}&q=smart%20watch"

# Initialize lists for storing data
all_brands = []
all_prices = []

# Loop through all pages (adjust range as needed)
for page in range(1, 5):  # Assuming 5 pages
    print(f"Scraping page {page}...")
    
    # Open the URL for the current page
    driver.get(base_url.format(page=page))
    driver.implicitly_wait(10)
    
    # Extract product elements
    product_elements = driver.find_elements(By.CLASS_NAME, "Bm3ON")
    price_elements = driver.find_elements(By.CLASS_NAME, "ooOxS")
    
    # Extract product names
    for element in product_elements:
        try:
            name_element = element.find_element(By.CLASS_NAME, "RfADt")
            all_brands.append(name_element.text.strip())
        except:
            all_brands.append(None)
    
    # Extract prices
    for element in price_elements:
        try:
            all_prices.append(element.text.strip())
        except:
            all_prices.append(None)
    
    # Add a small delay to avoid rate-limiting
    time.sleep(2)

# Quit the browser
driver.quit()

# Save all data to a pandas DataFrame
data = {
    "Product Name": all_brands,
    "Price": all_prices
}
df = pd.DataFrame(data)

# Export to CSV
csv_filename = "all_smart_watch_data.csv"
df.to_csv(csv_filename, index=False)

print(f"Data saved to {csv_filename}")
```

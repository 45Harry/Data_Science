```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import os

url = "https://www.daraz.com.np/catalog/?spm=a2a0e.pdp_revamp.search.3.5b9b4e336etQR4&q=iphone+15+pro+max&from_searchbox_hotwords=1&clickTrackInfo=matchType--99___wh_pid--___type--cli___scene--diwen___bucket--0___spm_id--hint.hp.sbox___result--iphone%2015%20pro%20max___textId--375550417040772256___score--0.748537___itemId--5___abtest--___pvid--66f7fb1c-6a7d-4493-bfcd-e105d3fcd0a3___scoreChain--0.000749,0.000000,0.000000___scm--1007.25945.379337.0___ifMannual--N"

url = "https://www.daraz.com.np/catalog/?spm=a2a0e.searchlist.search.6.7e398c4fkunTxi&q=smart+watch&from_searchbox_hotwords=1&clickTrackInfo=matchType--99___wh_pid--___type--cli___scene--diwen___bucket--0___spm_id--hint.hp.sbox___result--smart%20watch___textId--6319298025245409390___score--0.600801___itemId--5___abtest--___pvid--9890907b-b9cd-4a1c-aa6d-4bfadbcd9461___scoreChain--0.000601,0.000000,0.000000___scm--1007.25945.379337.0___ifMannual--N"
base_url = "https://www.daraz.com.np/catalog/?page={page}&q=lapto"

driver = webdriver.Firefox()
   
item_names = []
item_price =[]
#driver.title
for page in range(1,5):
    print(f"Scraping page {page}...")

    driver.get(base_url.format(page =page))
    driver.implicitly_wait(15)
    
    #locating itemname by class name
    item = driver.find_elements(By.CLASS_NAME,"Bm3ON")
    
 
    #finding poduct names
    for product_nm in item:
        name = product_nm.find_element(By.CLASS_NAME,'RfADt')
        item_names.append(name.text.strip())
        
    #locating pice by class name 
    
    #EXTRACT THE TEXT OF EACH PRICE
    for pric in item:
        item_p = pric.find_element(By.CLASS_NAME,"ooOxS")
        item_price.append(item_p.text.strip())
        #print(pric)
        #time.sleep(2)
        
    
#quit the browser
driver.quit()


#dataframe
data = {'Product Name':item_names,
        'Price':item_price}
df = pd.DataFrame(data)

#export to csv
df.to_csv("all_laptop.csv",index=False)

print:(f"Data saved :)")

```

```python
import pandas as pd
df = pd.read_csv(f'{path}/df_file.csv')
df.head()
```

```python
paragraphs = df['Text'][0]
```

```python
paragraphs
```

```python
import nltk 
```

```python
#sentences = paragraphs.apply(nltk.sent_tokenize)
sentences = nltk.sent_tokenize(paragraphs)
```

```python
#sentences = sentences
sentences
```

```python
type(sentences)
```

```python
# Applyging regular expressing to clean the unwanted character
import re
corpus = []
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z0-9]',' ',sentences[i])
    review = review.lower()
    corpus.append(review)   
```

```python
corpus
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
#applying Stemming 
stemmed_word = []
for i in corpus:
    words = nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            word = stemmer.stem(word)
            stemmed_word.append(word)
```

```python
stemmed_word
```

```python
#applying Lemmiatizer
lemmatized_word = []
for i in corpus:
    words = nltk.word_tokenize(i)
    for word in words:
        if word not  in set(stopwords.words('english')):
            word = lemmatizer.lemmatize(word)
            lemmatized_word.append(word)

```

```python
lemmatized_word
```

```python
# Applying in one shot
corpus = []
for i in range(len(sentences)):
    review = re.sub('^a-zA-Z0-9',' ',sentences[i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

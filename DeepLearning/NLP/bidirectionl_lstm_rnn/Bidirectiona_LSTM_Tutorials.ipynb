```python
# get the independetn tfeatures

X  = df.drop('label',axis=1)
y = df['label']
```

```python
from tensorflow.keras.layers import Embedding , LSTM , Dense , Bidirectional
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# vocabulary size
voc_size = 5000
```

```python
messages = X.copy()
```

```python
messages.reset_index(inplace=True)
messages.head()
```

```python
import nltk
import re
from nltk.corpus import stopwords

nltk.download('stopwords')
```

```python
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
corpus = []

for i in range(0,len(messages)):
  print(i)
  review = re.sub('[^a-zA-Z]',' ',messages['title'][i])
  review = review.lower()
  review = review.split()

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
one_hot_repr = [ one_hot(words,voc_size) for words in corpus]
print(one_hot_repr)
```

```python
## Embedding Representation

sent_length = 20
embedded_docs = pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
embedded_docs
```

```python
# Creating Model bidirectionla
embedding_vector_feature = 40


model = Sequential()
model.add(Embedding(voc_size,embedding_vector_feature,input_shape=( sent_length,)))

model.add(Bidirectional(LSTM(100)))

model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

print(model.summary())
```

```python
## convertion the embeded_docx to array to fit in the model
import numpy as np
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_final,y_final,test_size = 0.2,random_state=18,stratify=y_final)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
## Model Traning
model.fit(X_train,y_train,validation_data = (X_test,y_test),epochs=20,batch_size=32)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = np.where(y_pred>0.5,1,0)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score

confusion_matrix(y_test,y_pred)
```

```python
print(accuracy_score(y_test,y_pred))
```

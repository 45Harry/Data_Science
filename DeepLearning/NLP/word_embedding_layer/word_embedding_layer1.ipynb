# word Embedding Tchniques Using Embedding Layer in Keras

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    'the glass of mile',
    'the glass of  juice',
    'the cup of tea',
    'I am a good boy',
    'T am a Good Developer',
    'understanding the meaning of words',
    'your videos are good'
]
```

```python
## vocabulaze size 
voc_size = 500 # represents the total number of unique word in the entire sentences
```

### One HOt Representation

```python
onehot_repr = [one_hot(words,voc_size) for words in sent]
print(onehot_repr)
```

### Word Embedding Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential

import numpy as np
```

```python
sent_length = 8
embedded_docs = pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
n_features = 10 # vactotr of size 10 will be used to represent each index from embedded_docs
```

```python
model = Sequential()
model.add(Embedding(voc_size,10,input_shape=(sent_length,)))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
embedded_docs[0]
```

```python
print(model.predict(embedded_docs))
```

## assignment

```python
### assignment

sent =[
    'the world is a better place',
    'Batman series in my favourite movie',
    'I like DC movies',
    'the cat is eating the food',
    'tom and jerry is my favorite movie'
]
```

```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
```

```python
#voc_size
voc_size = 999
sent_one_hot = [one_hot(words,voc_size) for words in sent]
print(sent_one_hot)
```

```python
# padding 
sent_length = 10
n_features = 12
embedded_sents = pad_sequences(sent_one_hot,padding='pre',maxlen=sent_length)
print(embedded_sents)
```

```python
# word embedding using embedding layers

model = Sequential()
model.add(Embedding(voc_size,n_features,input_shape=(sent_length,)))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
embedded_sents[0]
```

```python
model.predict(embedded_sents)
```

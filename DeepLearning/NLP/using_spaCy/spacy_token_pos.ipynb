```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
with open('news_story.txt') as f:
    news_text = f.read()
```

```python
print(news_text)
```

```python
doc = nlp(news_text)

for token in doc:
    print(token , ' | ', token.tag_,' | ', spacy.explain(token.tag_))
    
```

```python
doc = nlp("Ram quits the job")

print(doc[1].text, "|", doc[1].tag_, "|", spacy.explain(doc[0].tag_))
```

```python
count  = doc.count_by(spacy.attrs.POS)
count
```

```python
for k, v in count.items():
    print(doc.vocab[k].text, '|',v)
```

```python
doc = nlp(news_text)
nouns_tokens = []
numberical_tokens = []
for token in doc:
    if token.tag_ in ['NN']:
        #print(token,' | ',token.tag_)
        nouns_tokens.append(token)
    elif token.pos_ == 'NUM':
        numberical_tokens.append(token)
```

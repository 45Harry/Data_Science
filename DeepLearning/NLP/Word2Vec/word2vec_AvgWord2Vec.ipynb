```python
#Project
import pandas as pd

message = pd.read_csv('SMSSpamCollection',sep='\t',names = ['label','message'])
```

```python
message
```

```python
message.shape
```

```python
# Data cleaning and preprocessing

import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0,len(message)):
  review = re.sub('[^a-z,A-Z]',' ',message['message'][i])
  review = review.lower()
  review = review.split()

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
# cratint eh bag of words model

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500)
X  = cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
X[0]
```

```python
y = pd.get_dummies(message['label'])
y=y.iloc[:,1].values
```

```python
# trian test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = 18)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train,y_train)
```

```python
#prediction
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_test,y_pred)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

```python
import gensim
import gensim.downloader as api

wv = api.load('word2vec-google-news-300')

```

```python

import nltk
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0,len(message)):
  review = re.sub('[^a-zA-Z]',' ',message['message'][i])
  review = review.lower()
  review = review.split()

  review = [lemmatizer.lemmatize(word) for word in review if word not in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess


```

```python
corpus[0]
```

```python
words = []
for sent in corpus:
  sent_token = sent_tokenize(sent)
  for sent in sent_token:
    words.append(simple_preprocess(sent))
```

```python
words
```

```python
# lets train word2vec from scratch

model = gensim.models.Word2Vec(words,window=5,min_count = 2) #was used
```

```python
import numpy as np
def avg_word2vec(doc):
  #remove out-of-vocabulary words
  #sent = [word for word in doc if word in model.wv.index_to_key]
  #print(sent)

  return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis = 0)
```

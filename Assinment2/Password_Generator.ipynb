```````````````````````python
import string
import random
```````````````````````

```````````````````````python

passLength = int(input("Enter the length of your desire password: "))
include_numbers = input("Want to add numbers in your password(y/n): ").lower() == 'y'  #inline if statement .. it will return bool value
include_special_chars = input("Want to add special Characters in your password(y/n): ").lower() == 'y'
include_uppercase = input("Want to add upper Character in your password(y/n)").lower() == 'y'
include_lowercase = input("Want to add lower Character in your password(y/n)").lower() == 'y'
```````````````````````

```````````````````````python
def generatePassword(passLength,include_numbers,include_special_chars,include_uppercase,lowChar):
    characters = '' # empty string varaible
    if include_numbers:
        characters += string.digits #randomaly add number character
    if include_special_chars:
        characters += string.punctuation #randomaly add special charcter
    if include_uppercase:
        characters += string.ascii_uppercase #randomaly add upper case 
    if include_lowercase:
        characters += string.ascii_lowercase # randomaly add lowercase 

    if not characters :
        return "Error: At least one character type must be selected>"

    password = ''.join(random.choice(characters) for _ in range(passLength)) #combine all the characters randomaly upto the given length
    return password
```````````````````````

```````````````````````python
password = generatePassword(passLength,include_numbers,include_special_chars,include_uppercase,include_lowercase)
print("Your generated password is: ",password)
```````````````````````

```python
total_bill = list(df['total_bill']) # assign the total_bill column to a variable called total_bill
```

```python
import numpy as np # import the numpy library
mean = np.mean(total_bill) # calculate the mean of the total_bill column
std = np.std(total_bill) # calculate the standard deviation of the total_bill column
```

```python
mean, std # display the mean and standard deviation
```

```python
normalized_data = [] # create an empty list called normalized_data
for i in total_bill: # loop through the total_bill column
    z_score = (i - mean) / std # calculate the z-score
    normalized_data.append(z_score) # append the z-score to the normalized_data list
```

```python
max(normalized_data) # display the maximum value in the normalized_data list
```

```python
min(normalized_data) # display the minimum value in the normalized_data list
```

```python
print(normalized_data) # display the normalized_data list
```

```python
sns.histplot(total_bill, kde=True) # create a histogram of the total_bill column
```

```python
sns.histplot(normalized_data, kde=True) # create a histogram of the normalized_data list
```

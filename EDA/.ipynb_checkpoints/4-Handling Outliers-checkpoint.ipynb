Outliers impact the machine learning algorithms which decreases the performance and accuracy of algorithms.

Outlier is important to check diseases like cancer.

5 Number Summary

- Minimum Value
- Q1 - 25 percentile
- Q2 - Median or 50 percentile
- Q3 - 75 Percentile
- Maximum Value

```python
import numpy as np # import numpy library as alias 'np'
```

```python
# create a list of numbers with some outliers
numbers = [45, 32, 56, 75, 89, 54, 32, 89, 90, 87, 67, 54, 45, 98, 99, 67, 74, 1000, 1100]
```

```python
np.percentile(numbers, 0)  # minimum value
```

```python
np.percentile(numbers, 25) # calculate the 25th percentile
```

```python
np.percentile(numbers, 50) # calculate the 50th percentile
```

```python
np.percentile(numbers, 75) # calculate the 75th percentile
```

```python
np.percentile(numbers, 100) # maximum value
```

Finding Outliers

[Lower Fence <--> Higher Fence]

```python
minimum, Q1, median, Q3, maximum = np.quantile(numbers, [0, 0.25, 0.5, 0.75, 1]) # calculate the five number summary
```

```python
maximum
```

```python
iqr = Q3 - Q1
print(iqr)
```

```python
lower_fence = Q1 - 1.5 * iqr
higher_fence = Q3 + 1.5 * iqr
```

```python
lower_fence, higher_fence
```

```python
numbers 
```

```python
outliers = []
for num in numbers:
    if num < lower_fence or num > higher_fence:
        outliers.append(num)
```

```python
print(outliers)
```
